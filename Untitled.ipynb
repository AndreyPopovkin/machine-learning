```python
import time

class Profiler:
	def __init__(self, cause = "something"):
		print "profile", cause

	def __enter__(self):
		self.time = time.clock()

	def __exit__(self, exc_type, exc_val, exc_tb):
		print 'done in %.7f s' % (time.clock() - self.time)
		return
```

```python
def fizzBuzz():
	for i in xrange(1, 101):
		if i % 5 == 0 and i % 3 == 0:
			print "FizzBuzz"
		elif i % 5 == 0:
			print "Buzz"
		elif i % 3 == 0:
			print "Fizz"
		else:
			print i

def multiplicate_matrix(mat1, mat2, h1, w2, same_dim):
	ans = []
	for i in xrange(h1):
		ans.append([])
		for q in xrange(w2):
			ans[i].append(0)	
			for k in xrange(same_dim):
				ans[i][q] += mat1[i][k] * mat2[k][q]
	return ans

def fib(n):
	ans = [[1], [0]]
	matrix = [
		[1, 1],
		[1, 0]
	]
	if n == 1: 
		return 0
	n -= 2
	for i in xrange(n):
		ans = multiplicate_matrix(matrix, ans, 2, 1, 2)
	return ans[0][0]

def sort(arr, l = -1, r = -1):
	if l == -1:
		l = 0
		r = len(arr) - 1
	if r - l < 1:
		return
	pivot = arr[l]
	left = l
	right = r
	while(left <= right):
		if arr[left] < pivot:
			left += 1
		elif arr[right] > pivot:
			right -= 1
		else:
			arr[left], arr[right] = arr[right], arr[left]
			left += 1
			right -= 1
	sort(arr, l, right)
	sort(arr, left, r)

def line_sort(arr):
	return (line_sort([x for x in arr if x < arr[0]]) + \
		[x for x in arr if x == arr[0]] + line_sort([x for x in arr if x > arr[0]])) if len(arr) > 1 else arr


fizzBuzz()

print fib(100)

arr = [1, -2, 3, -31, 42, 4543, -34, 2343]
sort(arr)
print arr

arr = [1, -2, 3, -31, 42, 4543, -34, 2343]
arr = line_sort(arr)
print arr
```

```python
from numpy import linspace
from numpy import ndarray

def my_linspace(start, stop, num = 50):
	ans = []
	#array('f')
	for i in xrange(num):
		ans.append(start + (stop - start) * i / (num - 1))
	return ans

def my_comprehension_linspace(start, stop, num = 50):
	return [start + (stop - start) * i / (num - 1) for i in xrange(num)]



def test_linspace():
	with Profiler("my_linspace(1, 100, 10000000)"):
		a = my_linspace(1, 100, 10000000)

	with Profiler("my_comprehension_linspace(1, 100, 10000000)"):
		a = my_comprehension_linspace(1, 100, 10000000)

	with Profiler("linspace(1, 100, 10000000)"):
		a = linspace(1, 100, 10000000)

test_linspace()

def cmp_array_with_list():
	with Profiler("init nparray"):
		arr = ndarray(shape=(1000000), dtype = float)
		for i in xrange(1000000):
			arr[i] = i

	with Profiler("init list"):
		lst = []
		for i in xrange(1000000):
			lst.append(i)

	with Profiler("sort nparray"):
		arr.sort()
		
	with Profiler("sort list"):
		lst.sort()

	with Profiler("reverce nparray"):
		for i in xrange(arr.size // 2):
			arr[i], arr[arr.size - i - 1] = arr[arr.size - i - 1], arr[i]

	with Profiler("reverce list"):
		lst.reverse()

	with Profiler("put items into nparray"):
		for i in xrange(1000000):
			arr[i * 10213 % arr.size] = i * 1224322

	with Profiler("put items into list"):
		for i in xrange(1000000):
			lst[i * 10213 % len(lst)] = i * 1224322

	ans = 0

	with Profiler("get items from nparray"):
		for i in xrange(1000000):
			ans += arr[i * 10217 % arr.size]

	ans = 0

	with Profiler("get items from nparray"):
		for i in xrange(1000000):
			ans += lst[i * 10217 % len(lst)]

cmp_array_with_list()
```

```python
import matplotlib.pyplot as plt
import numpy as np

lag = 0.0001
x = np.arange(1.0, 40+lag, lag)
y = np.add(np.cos(x), np.linspace(2, 2, 390001))
y2 = np.sqrt(x)
y3 = np.arctan(x)

fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.grid(True)
plt.plot(x, y, label='$cos(x)$', color='red')
plt.plot(x, y2, label='$sqrt(x)$', color='green')
plt.plot(x, y3, label='$log(x)$', color='blue')

plt.title('three elementary functions')
plt.xlabel('$OX$')
plt.ylabel('$OY$')
plt.legend()

fig.add_subplot(1, 2, 2)
plt.grid(True)
plt.plot(x, np.sin(x), label='$sin(x)$', color='green')
plt.plot(x, y3, label='$log(x)$', color='blue')

plt.title('two elementary functions')
plt.xlabel('$OX$')
plt.ylabel('$OY$')
plt.legend()

plt.savefig('two_plots.png', fmt='png')

plt.show()
```
